# Data wrangling task - extract job listings from text file

Environment: Python 3.8 and Jupyter notebook
Libraries used:
- re : For regular expression
- sys : For file operations
- json : For saving as JSON file

## 1. Introduction
This mini-project extracts job posting data from a text file. Regular expressions are used to separate the text data into individual job postings.<br>

Each job posting contains multiple known "keywords" which correspond to specific "tags" / "nodes". One regex was used to split raw text into job postings, and extract ID (identifier) of each job posting. A second regex was used to identify keywords, compare against a known set and text following each keyword is parsed using a FOR loop. The FOR loop also handled few keywords (tags) which needed to be structured as a list of sub-tags.<br>

The text thus pre-processed was stored in a dictionary, and then persisted into a XML file with appropriate tags and a JSON file (using json package).<br>

A total of 30704 job postings were parsed from input text file. Across these job postings, 58 different keywords were identified, which translated to 11 specific output tags in parsed data files (JSON & XML). Of the 11 output tags, 3 tags are nested and stored as list of sub-tags.

## 2. Import libraries

```python
import re
import sys
import json
import html
```

## 3. Open the file and examine data

### 3.1 Load text file

```python
# Open Job listing file
with open('joblistings.dat', 'r') as file:
    jobtext = file.read()
```

### 3.2 Examine text file

As a preparatory task,
- Regular expression <code><font color=red>r'^[\w]+[\s_]*[\w]*?:'</font></code> was used to extract list of all combinations of "raw" tags.
- From this list of 1064 raw tags, invalid "tags" were filtered out based on text following each in text file.
- The final list of 58 valid raw tags were mapped to corresponding "output tags" separately, and used in function to validate tags.

### 3.3 Split text into job postings

Package regex id used to find 30 consecutive "-" characters in the text, which is delimiter for job postings. Using this regex, first line containg tag "ID" is split as identifier of the job listing, and subsequent multiline text (till next delimiter) is stored as job content.<br>

Regex used = <code><font color=red>(ID[^\r\n]*)(.*?)(?:[-]{30})</font></code>. <br>
Flags applied for regex = <code>re.DOTALL|re.MULTILINE</code><br>

The regex splits the entire content into a list of 2 values in each item. Value 1 (Index <code>\[0\]</code>) is ID value, Value 2 (Index <code>\[1\]</code>) is remaining part of the job listing.<br>

**Run regex 1**<br>
This regex splits text into job postings and store in a 2D list

```python
joblist = []
regex = re.compile('(ID[^\r\n]*)(.*?)(?:[-]{30})',re.DOTALL|re.MULTILINE)
joblist = regex.findall(jobtext)
```

### 3.3 Generic functions used to process job content
Below code block contains 2 functions, which will be used later in the main code.

- <code>**validTag()**</code> - This function validates if an identified tag is valid & if valid, returns the corresponding output tag.
- <code>**splitTagVal()**</code> - This function splits a tag value into multiple sub-tags (child-tags). This is needed for *job_descriptions*, *job_responsibilities*, *required_qualifications* tags.

```python
# function to split Tag value into multiple sub tags. 
# input = concanated string
# Return = string list of individual sub-tags
def splitTagVal(sLine: str):
    sVal = []
    sLine = sLine.strip()
    if sLine == '':
        sVal = ['N/A']
    else:
        # if first line does not have a "-", add so that it is considered as a child-tag.
        # Few job posting in text file has this issue. 
        if sLine[0] != '-':
            sLine = '- ' + sLine
        # Replace start of sub-tag with "~" to ensure any in-line "-" is picked up as-is, 
        # and is not confused to be a child-tag
        sLine = sLine.replace('- ','~')
        
        # Now convert the entire text into a single line - replace \n into <space>
        sLine = ' '.join(sLine.splitlines())
        #split into list by delimiter of "~ "
        rg = re.compile(r'[~]([\w\s.;,/@&":()+-]+)',re.DOTALL|re.MULTILINE)
        sLst = rg.findall(sLine)
        # If multiple child-tag lines are found, add as items into result list
        if len(sLst) > 0:
            for sItm in sLst:
                sVal.append(sItm.strip().lstrip('~'))
        else:
        # If no child-tags were found, just pass back the input variable as one child-tag
            sVal.append(sLine)
    return sVal

# function to check if identified tag is valid. 
# input = tag name from text file
# Return = valid tag name to be put into XML / JSON file. If invalid, returns blank
def validTag(sTg: str):
    sValidTag = ''
    sTg = sTg.lower().strip()
    if sTg in ['_info', 'about', 'about company', 'about_company', 'companys_info']:
        sValidTag = 'about_company'
    elif sTg in ['application_deadl', 'application_dl', 'dead_line', 'deadline', 'deadlines']:
        sValidTag = 'application_deadline'
    elif sTg in ['job_proc', 'job_procs', 'procedure', 'procedures']:
        sValidTag = 'application_procedure'
    elif sTg in ['_description',  'description', 'job description', 'job_desc']:
        sValidTag = 'job_descriptions'
    elif sTg in ['job responsibilities', 'job_resps', 'resp', 'responsibilities', 'responsibility']:
        sValidTag = 'job_responsibilities'
    elif sTg in ['_loc', '_locs', 'job_loc', 'location', 'locations']:
        sValidTag = 'location'
    elif sTg in ['qualification', 'qualifications', 'qualifs', 'req_quals', 'required qualifications']:
        sValidTag = 'required_qualifications'
    elif sTg in ['compensation details', 'job_sal', 'remuneration', 'salary']:
        sValidTag = 'salary'
    elif sTg in ['date_start', 'dates', 'start date', 'start_da', 'start_date', 'starting date']:
        sValidTag = 'start_date'
    elif sTg in ['_ttl', 'job title', 'job_t', 'title', 'titles']:
        sValidTag = 'title'
    else:
        sValidTag = ''
    
    return sValidTag
```

### 3.4 Process each job posting, split into tags and store in a dictionary
Below is main part of the code. Approach is to take the list "joblist", created by first regex and pass it through a series of functions, each of which will further break up the job listing into individual listing & its corresponding tags / nodes.

This code block utilises a FOR loop to traverse through each job listing. In a nested loop, each line of the job listing data is analysed for any available tags, and based on sequence of valid tags, categorised into relevant tags for output to file (XML & JSON)

** *Algorithm used:* **
1. For each job listing,
2. Declare job listing container (blank) dictionary
3. Store "ID" from job listing Column[0] into dictionary
4. For each line in job listing context "Column[1]",
5. Check if line has a "tag" as per regex. 
6. If tag is found, validate tag.
7. If valid tag, store previous content of Tag & value into appropriate tag in dictionary.
8. If Valid tag, initialise Tag & value variables. process line
9. If not valid tag, process line as part of current Tag & Value.
10. When all lines are complete, go through the dictionary tags, identify if anything is blank and fill default value "N/A".
11. Add dictionary into Main dictionary list "jobListDict"

```python
#Initialise all variables
sId = ''
sJobDet = []
tg = ''
sTg = ''
sTag = ''
sTagVal = []
sVal = []

#Initialise dictionaries and lists used
#Dictionary to store list of all job listings
jobListDict = {}
jobListDict['listing'] = []
#Dictionary variable to store each job listing
jobDict = {}

#Function to add tag into dictionary.
#Based on tag, this function stores either a free-form text or list of child-tags
def addTag(sName: str, val: list):
    #If tag needs child-tags, split content into child-tags
    if sName in ["job_descriptions","job_responsibilities","required_qualifications"]:
        sVal = splitTagVal(' '.join(val).strip())
        
        #Store child tags under appropriate child-tag name
        if sName == "job_descriptions":
            jobDict[sName]['description'] = sVal
        elif sName == "job_responsibilities":
            jobDict[sName]['responsibility'] = sVal
        elif sName == "required_qualifications":
            jobDict[sName]['qualification'] = sVal
    else:
        #If tag does not need child tag, add content as-is
        jobDict[sName] = (' '.join(val)).strip()

#Main code section. Contains nested FOR loops

# for each job in "joblist"
for job in joblist:

#initialise job listing specific variables
    tg = ''
    sId = ''
    sJobDet = []
    sTg = ''
    sTag = ''
    sTagVal = []
    sVal = []
    
    # Define default structure of jobDict. This ensures no tags are missed later.
    jobDict = {
        "_id": "",
        "title": "",
        "location": "",
        "job_descriptions": {"description": []},
        "job_responsibilities": {"responsibility": []},
        "required_qualifications": {"qualification": []},
        "salary": "",
        "application_procedure": "",
        "start_date": "",
        "application_deadline": "",
        "about_company": ""
    }
    
    #Get ID
    sId = job[0].lstrip('ID:').strip()
    
    #Get job details
    sJobDet = job[1].splitlines()
    jobDict['_id'] = sId
    
    #Traverse through each line in Job details (content)
    for jobline in sJobDet:
        jobline = jobline.strip()
        
        # If a tag is found at beginning of line
        if re.match(r'^[\w]+[\s_]?[\w]*?:', jobline):
            #Extract tag
            tg = re.findall(r'^[\w]+[\s_]*[\w]*?:', jobline)[0].split(':')[0]
            
            # Get validated tag name. If empty, this is an invalid tag
            sTg = validTag(tg)
            
            #If this is not the first tag, ensure old tag is stored before processing this line
            if sTg != '':
                #store old tag first
                if sTag != '':
                    addTag(sTag, sTagVal)
                    # after storing, reset values and initialise
                    sTag = ''
                    sVal = []
                    sTagVal = []
                
                #variable to store into dictionary
                sTag = sTg
                #extract portion of line (with tag) after the tag, this should be added to tag value
                sTagVal.append(jobline.replace(tg+':','',1).strip())
            
            # If this is not a valid tag. Just continue with earlier tag
            elif sTag != '':
                sTagVal.append(jobline)
        
        # If no tag is found, this should be added to earlier tag value
        elif jobline != '' and sTag != '':
            sTagVal.append(jobline)
    
    #add the last tag & value to jobDict
    if sTag != '':
        addTag(sTag, sTagVal)

# Set default values if a tag is blank or empty
    if jobDict['about_company'].strip() == '':
        jobDict['about_company'] = 'N/A'
        
    if jobDict['application_deadline'].strip() == '':
        jobDict['application_deadline'] = 'N/A'
        
    if len(jobDict['required_qualifications']['qualification']) == 0:
        jobDict['required_qualifications']['qualification'] = ['N/A']
        
    if jobDict['location'].strip() == '':
        jobDict['location'] = 'N/A'
        
    if len(jobDict['job_descriptions']['description']) == 0:
        jobDict['job_descriptions']['description'] = ['N/A']
        
    if len(jobDict['job_responsibilities']['responsibility']) == 0:
        jobDict['job_responsibilities']['responsibility'] = ['N/A']
        
    if jobDict['salary'].strip() == '':
        jobDict['salary'] = 'N/A'
        
    if jobDict['start_date'].strip() == '':
        jobDict['start_date'] = 'N/A'
        
    if jobDict['application_procedure'].strip() == '':
        jobDict['application_procedure'] = 'N/A'
    
# Add this job listing to list "listing"
    jobListDict['listing'].append(jobDict)
```

### 3.5 Persist content into XML and JSON files

```python

# Declare final JSON object to persist in file
jFinalDict = {}
jFinalDict['listings'] = jobListDict

# Use json package to save dictionary as JSON file.
# speficy indenting to 4 characters to make it easy to read
with open('joblistings.json', 'w') as outfile:
    json.dump(jFinalDict, outfile, indent = 4)

# Use FOR loop to store dictionary list into XML file.
# This uses format specified in "Sample.XML" file
with open('joblistings.xml', 'w') as xmlfile:
    xmlfile.write('<?xml version="1.0" encoding="UTF-8" ?>\n')
    xmlfile.write('<listings>\n')
    for jobDict in jobListDict['listing']:
        xmlfile.write(' ' * 2 + "<listing id='" + html.escape(jobDict["_id"]) + "'>\n")
        xmlfile.write(' ' * 4 + '<title>' + html.escape(jobDict['title']) + '</title>\n')
        xmlfile.write(' ' * 4 + '<location>' + html.escape(jobDict['location']) + '</location>\n')

        xmlfile.write(' ' * 4 + '<job_descriptions>\n')
        for sTagVal in jobDict['job_descriptions']['description']:
            xmlfile.write(' ' * 6 + '<description>\n')
            xmlfile.write(' ' * 6 + html.escape(sTagVal) + '\n')
            xmlfile.write(' ' * 6 + '</description>\n')
        xmlfile.write(' ' * 4 + '</job_descriptions>\n')

        xmlfile.write(' ' * 4 + '<job_responsibilities>\n')
        for sTagVal in jobDict['job_responsibilities']['responsibility']:
            xmlfile.write(' ' * 6 + '<responsibility>\n')
            xmlfile.write(' ' * 6 + html.escape(sTagVal) +'\n')
            xmlfile.write(' ' * 6 + '</responsibility>\n')
        xmlfile.write(' ' * 4 + '</job_responsibilities>\n')

        xmlfile.write(' ' * 4 + '<required_qualifications>\n')
        for sTagVal in jobDict['required_qualifications']['qualification']:
            xmlfile.write(' ' * 6 + '<qualification>\n')
            xmlfile.write(' ' * 6 + html.escape(sTagVal) + '\n')
            xmlfile.write(' ' * 6 + '</qualification>\n')
        xmlfile.write(' ' * 4 + '</required_qualifications>\n')

        xmlfile.write(' ' * 4 + '<salary>' + html.escape(jobDict['salary']) + '</salary>\n')
        xmlfile.write(' ' * 4 + '<application_procedure>' + html.escape(jobDict['application_procedure']) + '</application_procedure>\n')
        xmlfile.write(' ' * 4 + '<start_date>' + html.escape(jobDict['start_date']) + '</start_date>\n')
        xmlfile.write(' ' * 4 + '<application_deadline>' + html.escape(jobDict['application_deadline']) + '</application_deadline>\n')
        xmlfile.write(' ' * 4 + '<about_company>' + html.escape(jobDict['about_company']) + '</about_company>\n')

        xmlfile.write('  </listing>\n')

    xmlfile.write('</listings>\n')    
    xmlfile.close()
```
